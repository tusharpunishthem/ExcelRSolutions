1. What is Normalization & Standardization and how is it helpful?
Normalization and Standardization are two common techniques used to scale features in a dataset, especially when preparing data for machine learning algorithms.

Normalization:
Definition: Normalization (also known as Min-Max scaling) rescales the feature values to a fixed range, usually [0, 1]. The formula for normalization is:

[ X' = \frac{X - X_{min}}{X_{max} - X_{min}} ]

Use Case: Normalization is particularly useful when the features have different units or scales. For example, if one feature is measured in kilometers and another in meters, normalization ensures that both features contribute equally to the distance calculations in algorithms like K-Nearest Neighbors (KNN) or when using gradient descent.

Benefits:

Helps in speeding up convergence in optimization algorithms.
Prevents features with larger ranges from dominating the model.
Standardization:
Definition: Standardization (also known as Z-score normalization) transforms the data to have a mean of 0 and a standard deviation of 1. The formula for standardization is:

[ X' = \frac{X - \mu}{\sigma} ]

where ( \mu ) is the mean of the feature and ( \sigma ) is the standard deviation.

Use Case: Standardization is useful when the data follows a Gaussian distribution (bell curve). It is commonly used in algorithms that assume normally distributed data, such as linear regression, logistic regression, and principal component analysis (PCA).

Benefits:

Makes the data more interpretable and comparable.
Helps in improving the performance of algorithms that rely on distance calculations.
2. What techniques can be used to address multicollinearity in multiple linear regression?
Multicollinearity occurs when two or more independent variables in a regression model are highly correlated, which can lead to unreliable coefficient estimates and inflated standard errors. Here are several techniques to address multicollinearity:

1. Variance Inflation Factor (VIF):
Definition: VIF quantifies how much the variance of a regression coefficient is increased due to multicollinearity. A VIF value greater than 10 (or sometimes 5) indicates high multicollinearity.
Action: Remove or combine features with high VIF values.
2. Remove Highly Correlated Features:
Definition: Identify pairs of features with high correlation coefficients (e.g., above 0.8 or 0.9) and consider removing one of the features from the model.
Action: Use a correlation matrix to identify and drop one of the correlated features.
3. Principal Component Analysis (PCA):
Definition: PCA is a dimensionality reduction technique that transforms correlated features into a set of uncorrelated components.
Action: Use the principal components as predictors in the regression model instead of the original correlated features.
4. Regularization Techniques:
Definition: Regularization methods like Lasso (L1 regularization) and Ridge (L2 regularization) can help mitigate the effects of multicollinearity by adding a penalty to the loss function.
Action: Use Lasso to shrink some coefficients to zero, effectively performing feature selection, or use Ridge to reduce the impact of multicollinearity without eliminating features.
5. Combine Features:
Definition: Create a new feature that combines the correlated features (e.g., taking the average or sum).
Action: This reduces the number of features while retaining the information.
6. Increase Sample Size:
Definition: Sometimes, multicollinearity can be reduced by increasing the sample size, which can provide more information and reduce the correlation between features.
Action: Collect more data if feasible.
